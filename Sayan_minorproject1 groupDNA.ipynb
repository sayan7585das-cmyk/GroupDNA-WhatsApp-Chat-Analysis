## GroupDNA

WhatsApp Group Chat Analyzer

```python
import numpy as np
from datetime import datetime
```

## Load the Chat File

```python
with open("groupDNA_project/hostel_bois.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

print(len(lines))
```

## Preview the Dataset

```python
print(lines[:5])
```

 Parse 


```python
messages = []

for line in lines:
    line = line.strip()

    try:
        date_time, rest = line.split(" - ", 1)

        if ":" not in rest:
            continue

        sender, message = rest.split(":", 1)

        messages.append({
            "timestamp": date_time.strip(),
            "sender": sender.strip(),
            "message": message.strip()
        })

    except:
        continue
```

## Verify Parsed Messages

```python
print("Total parsed messages:", len(messages))
print(messages[0])
```

## Basic Statistics



```python
message_count = {}

for msg in messages:
    sender = msg["sender"]

    if sender not in message_count:
        message_count[sender] = 0

    message_count[sender] += 1
```

```python
most_active = max(message_count, key=message_count.get)

print("Most Active Member:", most_active)
print("Messages:", message_count[most_active])
```

```python
least_active = min(message_count, key=message_count.get)

print("Least Active Member:", least_active)
print("Messages:", message_count[least_active])
```

# Rendering the Activity Heatmap

```python
print("=" * 80)
print("ACTIVITY HEATMAP (Messages by Hour)")
print("=" * 80)

hours = [0, 3, 6, 9, 12, 15, 18, 21]

print(f"{'Participant':<10}", end=" ")

for h in hours:
    print(f"{h:02d}", end=" ")

print()

for i, person in enumerate(participants):

    row = activity_matrix[i]

    max_value = np.max(row)

    print(f"{person:<10}", end=" ")

    for h in hours:

        value = row[h]

        if value == 0:
            symbol = "."

        else:

            ratio = value / max_value

            if ratio <= 0.25:
                symbol = "."

            elif ratio <= 0.50:
                symbol = "░"

            elif ratio <= 0.75:
                symbol = "▒"

            else:
                symbol = "█"

        print(symbol, end="  ")

    print()
```

# Word Frequency Analysis

```python
stop_words = {
    "the","is","am","are","was","were","be","been","being",
    "a","an","and","or","to","of","in","on","at","for","with","from","by",
    "i","me","my","mine","you","your","yours","we","our","ours",
    "he","his","him","she","her","hers","they","them","their","theirs",
    "it","its","this","that","these","those",
    "have","has","had","do","does","did",
    "can","could","will","would","shall","should",
    "may","might","must",
    "how","what","when","where","which","why","who","whom",
    "so","about","just","today","everyone","yes","no","not",
    "ok","okay","haan","hai","na","ka","ki","ke"
}
```

```python
word_count = {}

for msg in messages:

    text = msg["message"].lower()

    # Remove common punctuation
    for ch in ".,!?;:()[]{}\"'":
        text = text.replace(ch, "")

    words = text.split()

    for word in words:

        if word in stop_words:
            continue

        if word == "<media":
            continue

        if word == "omitted>":
            continue

        if word not in word_count:
            word_count[word] = 0

        word_count[word] += 1
```

```python
sorted_words = sorted(
    word_count.items(),
    key=lambda x: x[1],
    reverse=True
)
```

```python
print("Top 10 Most Frequent Words")
print("-" * 40)

for word, count in sorted_words[:10]:
    print(f"{word:<15} {count}")
```

```python
print("Top 10 Words (Bar Chart)")
print("-" * 40)

top10 = sorted_words[:10]

max_count = top10[0][1]

for word, count in top10:

    bar_length = int((count / max_count) * 30)

    print(f"{word:<15} {'█' * bar_length} {count}")
```

# Response Pattern Analysis


```python
response_times = {}

for person in participants:
    response_times[person] = []
```

```python
for i in range(1, len(messages)):

    current = messages[i]
    previous = messages[i - 1]

    # Ignore consecutive messages from the same person
    if current["sender"] == previous["sender"]:
        continue

    gap = current["datetime"] - previous["datetime"]

    minutes = gap.total_seconds() / 60

    response_times[current["sender"]].append(minutes)
```

```python
average_response = {}

for person in participants:

    if len(response_times[person]) > 0:

        average_response[person] = (
            sum(response_times[person]) /
            len(response_times[person])
        )

    else:

        average_response[person] = 0
```

```python
print("AVERAGE RESPONSE TIMES")
print("=" * 40)

for person in participants:

    avg = average_response[person]

    if avg < 60:
        print(f"{person:<10} {avg:.1f} minutes")

    else:
        print(f"{person:<10} {avg/60:.1f} hours")
```

```python
fastest = min(average_response, key=average_response.get)
slowest = max(average_response, key=average_response.get)

print("\nRESPONSE SUMMARY")
print("=" * 40)

print(f"Fastest Replier : {fastest} ({average_response[fastest]:.1f} minutes)")

if average_response[slowest] < 60:
    print(f"Slowest Replier : {slowest} ({average_response[slowest]:.1f} minutes)")
else:
    print(f"Slowest Replier : {slowest} ({average_response[slowest]/60:.1f} hours)")
```

# Longest Silent Streak

```python
active_days = {}

for person in participants:
    active_days[person] = set()

for msg in messages:
    active_days[msg["sender"]].add(msg["datetime"].date())
```

```python
from datetime import timedelta

all_days = []

current = first_date.date()

while current <= last_date.date():
    all_days.append(current)
    current += timedelta(days=1)
```

```python
silent_streak = {}

for person in participants:

    longest = 0
    current = 0

    for day in all_days:

        if day in active_days[person]:
            current = 0
        else:
            current += 1

            if current > longest:
                longest = current

    silent_streak[person] = longest
```

```python
print("LONGEST SILENT STREAKS")
print("=" * 40)

for person, days in sorted(
        silent_streak.items(),
        key=lambda x: x[1],
        reverse=True):

    print(f"{person:<10} {days} days")
```

# Personality Archetype Detection

```python
#Night Activity %
night_percentage = {}

for i, person in enumerate(participants):

    total = np.sum(activity_matrix[i])

    night = sum(activity_matrix[i][h] for h in [23,0,1,2,3,4])

    night_percentage[person] = (night / total) * 100
```

```python
#media count 
media_count = {}

for person in participants:
    media_count[person] = 0

for msg in messages:

    if msg["message"].strip().lower() == "<media omitted>":

        media_count[msg["sender"]] += 1
```

```python
#averag massage lenght
average_length = {}

for person in participants:

    total_words = 0
    total_messages = 0

    for msg in messages:

        if msg["sender"] != person:
            continue

        total_messages += 1
        total_words += len(msg["message"].split())

    average_length[person] = total_words / total_messages
```

```python
#consicutive massage burst
burst_score = {}

for person in participants:

    streaks = []

    current = 0

    for msg in messages:

        if msg["sender"] == person:

            current += 1

        else:

            if current > 0:
                streaks.append(current)

            current = 0

    if current > 0:
        streaks.append(current)

    if len(streaks) == 0:
        burst_score[person] = 0

    else:

        burst_score[person] = sum(streaks)/len(streaks)
```

```python
#caring words
care_words = {
    "sleep","eat","drink","water","study",
    "care","take","rest","medicine",
    "safe","home","please","dont","don't"
}

care_score = {}

for person in participants:

    score = 0

    for msg in messages:

        if msg["sender"] != person:
            continue

        text = msg["message"].lower()

        for word in care_words:

            if word in text:
                score += 1

    care_score[person] = score
```

```python
#caps massage
caps_score = {}

for person in participants:

    score = 0

    for msg in messages:

        if msg["sender"] != person:
            continue

        for word in msg["message"].split():

            if len(word) > 2 and word.isupper():

                score += 1

    caps_score[person] = score
```

```python
print("Night %")
print(night_percentage)

print("\nMedia")
print(media_count)

print("\nAverage Length")
print(average_length)

print("\nBurst")
print(burst_score)

print("\nCare")
print(care_score)

print("\nCaps")
print(caps_score)

print("\nSilent")
print(silent_streak)
```

```python
night_owl = max(night_percentage, key=night_percentage.get)

ghost = max(silent_streak, key=silent_streak.get)

spammer = max(burst_score, key=burst_score.get)

group_mom = max(care_score, key=care_score.get)

drama_queen = max(caps_score, key=caps_score.get)

meme_machine = max(media_count, key=media_count.get)

wordy_person = max(average_length, key=average_length.get)

fast_replier = min(average_response, key=average_response.get)
```

```python
archetypes = {}

for person in participants:
    archetypes[person] = []
```

```python
archetypes[night_owl].append("🌙 Night Owl")

archetypes[ghost].append("👻 Ghost")

archetypes[spammer].append("💬 Spammer")

archetypes[group_mom].append("❤️ Group Mom")

archetypes[drama_queen].append("😱 Drama Queen")

archetypes[meme_machine].append("😂 Meme Machine")

archetypes[wordy_person].append("📖 Wordy Person")

archetypes[fast_replier].append("⚡ Fast Replier")
```

```python
for person in participants:

    if len(archetypes[person]) == 0:

        archetypes[person].append("🙂 Regular Member")
```

```python
print("=" * 60)
print("PERSONALITY ARCHETYPES")
print("=" * 60)

for person in participants:

    print(f"\n{person}")

    for role in archetypes[person]:
        print("   ", role)
```
